```python
import numpy as np
import pandas as pd
```

```python
# read data
df = pd.read_csv('df.csv')
df.head()
```

```python
# Create dataframe
seaports = pd.DataFrame(df['Destination seaport code'].unique(), columns=['seaport code'])
seaports['place'] = np.nan
seaports['width'] = np.nan
seaports['height'] = np.nan
seaports
```

```python
seaports_locations = {'PLDAR': {'place': 'Darłowo', 'width': 54.4315, 'height': 16.3815},
                      'PLGDN': {'place': 'Gdańsk', 'width': 54.37815, 'height': 18.67575},
                      'PLNOW': {'place': 'Nowy Port, Gdańsk', 'width': 54.407140, 'height': 18.657409},
                      'PLGDY': {'place': 'Gdynia', 'width': 54.53205, 'height': 18.5264},
                      'PLPLC': {'place': 'Police', 'width': 53.56252, 'height': 14.588815},
                      'PLSWI': {'place': 'Świnoujście', 'width': 53.9095, 'height': 14.2785},
                      'PLSZZ': {'place': 'Szczecin', 'width': 53.441, 'height': 14.595},
                      'PLUST': {'place': 'Ustka', 'width': 54.585, 'height': 16.8545},
                      'PLKOL': {'place': 'Kołobrzeg', 'width': 54.182785, 'height': 15.55687}}
```

```python
# Fill dataframe
for seaport in seaports['seaport code'].unique():
    seaports['place'][seaports['seaport code'] == seaport] = seaports_locations[seaport]['place']
    seaports['width'][seaports['seaport code'] == seaport] = seaports_locations[seaport]['width']
    seaports['height'][seaports['seaport code'] == seaport] = seaports_locations[seaport]['height']
```
